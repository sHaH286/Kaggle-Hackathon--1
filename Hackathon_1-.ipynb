```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train_2= pd.read_csv('Surveydata_train.csv')
df_test_2 = pd.read_csv('Surveydata_test.csv')
```

```python
df_train_1 = pd.read_csv('Traveldata_train.csv')
df_test_1 = pd.read_csv('Traveldata_test.csv')
```

```python
df_train = pd.merge(df_train_1,df_train_2)
```

```python
df_test = pd.merge(df_test_1, df_test_2)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train['Online_support'] = df_train['Online_support'].replace('extremely poor', 'poor')
df_train['Platform_location'] = df_train['Platform_location'].replace('very inconvinient', 'Inconvinient')
df_train['Checkin_service'] = df_train['Checkin_service'].replace('extremely poor','poor')
df_train['Cleanliness'] = df_train['Cleanliness'].replace('extremely poor','poor')
df_train['Onboard_service'] = df_train['Onboard_service'].replace('extremely poor', 'poor')
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train.loc[df_train['Seat_comfort'].isnull() == True,'Seat_comfort'] = df_train['Seat_comfort'].mode()[0]
df_test.loc[df_test['Seat_comfort'].isnull() == True,'Seat_comfort'] = df_train['Seat_comfort'].mode()[0]

df_train.loc[df_train['Arrival_time_convenient'].isnull() == True,'Arrival_time_convenient'] = df_train['Arrival_time_convenient'].mode()[0]
df_test.loc[df_test['Arrival_time_convenient'].isnull() == True,'Arrival_time_convenient'] = df_train['Arrival_time_convenient'].mode()[0]

df_train.loc[df_train['Catering'].isnull() == True,'Catering'] = df_train['Catering'].mode()[0]
df_test.loc[df_test['Catering'].isnull() == True,'Catering'] = df_train['Catering'].mode()[0]

df_train.loc[df_train['Platform_location'].isnull() == True,'Platform_location'] = df_train['Platform_location'].mode()[0]
df_test.loc[df_test['Platform_location'].isnull() == True,'Platform_location'] = df_train['Platform_location'].mode()[0]

df_train.loc[df_train['Onboardwifi_service'].isnull() == True,'Onboardwifi_service'] = df_train['Onboardwifi_service'].mode()[0]
df_test.loc[df_test['Onboardwifi_service'].isnull() == True,'Onboardwifi_service'] = df_train['Onboardwifi_service'].mode()[0]

df_train.loc[df_train['Onboard_entertainment'].isnull() == True,'Onboard_entertainment'] = df_train['Seat_comfort'].mode()[0]
df_test.loc[df_test['Onboard_entertainment'].isnull() == True,'Onboard_entertainment'] = df_train['Seat_comfort'].mode()[0]

df_train.loc[df_train['Online_support'].isnull() == True,'Online_support'] = df_train['Online_support'].mode()[0]
df_test.loc[df_test['Online_support'].isnull() == True,'Online_support'] = df_train['Online_support'].mode()[0]

df_train.loc[df_train['Onlinebooking_Ease'].isnull() == True,'Onlinebooking_Ease'] = df_train['Onlinebooking_Ease'].mode()[0]
df_test.loc[df_test['Onlinebooking_Ease'].isnull() == True,'Onlinebooking_Ease'] = df_train['Onlinebooking_Ease'].mode()[0]

df_train.loc[df_train['Onboard_service'].isnull() == True,'Onboard_service'] = df_train['Seat_comfort'].mode()[0]
df_test.loc[df_test['Onboard_service'].isnull() == True,'Onboard_service'] = df_train['Seat_comfort'].mode()[0]

df_train.loc[df_train['Leg_room'].isnull() == True,'Leg_room'] = df_train['Leg_room'].mode()[0]
df_test.loc[df_test['Leg_room'].isnull() == True,'Leg_room'] = df_train['Leg_room'].mode()[0]

df_train.loc[df_train['Baggage_handling'].isnull() == True,'Baggage_handling'] = df_train['Baggage_handling'].mode()[0]
df_test.loc[df_test['Baggage_handling'].isnull() == True,'Baggage_handling'] = df_train['Baggage_handling'].mode()[0]

df_train.loc[df_train['Checkin_service'].isnull() == True,'Checkin_service'] = df_train['Checkin_service'].mode()[0]
df_test.loc[df_test['Checkin_service'].isnull() == True,'Checkin_service'] = df_train['Checkin_service'].mode()[0]

df_train.loc[df_train['Cleanliness'].isnull() == True,'Cleanliness'] = df_train['Cleanliness'].mode()[0]
df_test.loc[df_test['Cleanliness'].isnull() == True,'Cleanliness'] = df_train['Cleanliness'].mode()[0]

df_train.loc[df_train['Online_boarding'].isnull() == True,'Online_boarding'] = df_train['Online_boarding'].mode()[0]
df_test.loc[df_test['Online_boarding'].isnull() == True,'Online_boarding'] = df_train['Online_boarding'].mode()[0]

df_train.loc[df_train['Gender'].isnull() == True,'Gender'] = df_train['Gender'].mode()[0]
df_test.loc[df_test['Gender'].isnull() == True,'Gender'] = df_train['Gender'].mode()[0]

df_train.loc[df_train['CustomerType'].isnull() == True,'CustomerType'] = df_train['CustomerType'].mode()[0]
df_test.loc[df_test['CustomerType'].isnull() == True,'CustomerType'] = df_train['CustomerType'].mode()[0]

df_train.loc[df_train['Age'].isnull() == True,'Age'] = df_train['Age'].mode()[0]
df_test.loc[df_test['Age'].isnull() == True,'Age'] = df_train['Age'].mode()[0]

df_train.loc[df_train['TypeTravel'].isnull() == True,'TypeTravel'] = df_train['TypeTravel'].mode()[0]
df_test.loc[df_test['TypeTravel'].isnull() == True,'TypeTravel'] = df_train['TypeTravel'].mode()[0]
```

```python
df_train['DepartureDelay_in_Mins'].replace(np.nan, df_train['DepartureDelay_in_Mins'].median(),inplace=True)
df_test['DepartureDelay_in_Mins'].replace(np.nan, df_test['DepartureDelay_in_Mins'].median(),inplace=True)

df_train['ArrivalDelay_in_Mins'].replace(np.nan, df_train['ArrivalDelay_in_Mins'].median(),inplace=True)
df_test['ArrivalDelay_in_Mins'].replace(np.nan, df_test['ArrivalDelay_in_Mins'].median(),inplace=True)
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train.drop('ID',axis = 1,inplace = True)
test_file_ids = df_test['ID'] ## necessary step
df_test.drop('ID',axis = 1,inplace = True)
```

```python
df_train = pd.get_dummies(df_train,drop_first = True)
df_train.head()
```

```python
df_test = pd.get_dummies(df_test,drop_first = True)
df_test.head()
```

```python
cols = df_train.columns
```

```python
X = df_train.drop(['Overall_Experience'],axis = 1)
y = df_train['Overall_Experience']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(X,y,test_size = 0.3,random_state = 2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion='gini',max_depth=150,random_state=1)
dt_model.fit(X_train,y_train)

y_pred_train = dt_model.predict(X_train)
y_pred_val = dt_model.predict(X_validation)
```

```python
from sklearn import metrics
dt_acc_train = metrics.accuracy_score(y_train,y_pred_train)
dt_acc_val = metrics.accuracy_score(y_validation,y_pred_val)
print("Train Accuracy:",dt_acc_train)
print("Validation Accuracy:",dt_acc_val)
print()

dt_roc_train = metrics.roc_auc_score(y_train,y_pred_train)
dt_roc_val = metrics.roc_auc_score(y_validation,y_pred_val)
print("Train ROC AUC score:",dt_roc_train)
print("Validation ROC AUC score:",dt_roc_val)
print()

dt_clr_train = metrics.classification_report(y_train,y_pred_train)
dt_clr_val = metrics.classification_report(y_validation,y_pred_val)
print("Train Classification Report:\n",dt_clr_train)
print("Validation Classification Report:\n",dt_clr_val)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=10,criterion='gini',max_depth=150,random_state=1)
rf_model.fit(X_train,y_train)

y_pred_train = rf_model.predict(X_train)
y_pred_val = rf_model.predict(X_validation)
```

```python
from sklearn import metrics
dt_acc_train = metrics.accuracy_score(y_train,y_pred_train)
dt_acc_val = metrics.accuracy_score(y_validation,y_pred_val)
print("Train Accuracy:",dt_acc_train)
print("Validation Accuracy:",dt_acc_val)
print()

dt_roc_train = metrics.roc_auc_score(y_train,y_pred_train)
dt_roc_val = metrics.roc_auc_score(y_validation,y_pred_val)
print("Train ROC AUC score:",dt_roc_train)
print("Validation ROC AUC score:",dt_roc_val)
print()

dt_clr_train = metrics.classification_report(y_train,y_pred_train)
dt_clr_val = metrics.classification_report(y_validation,y_pred_val)
print("Train Classification Report:\n",dt_clr_train)
print("Validation Classification Report:\n",dt_clr_val)
```

```python
final_predictions = rf_model.predict(df_test)
submission=pd.DataFrame([test_file_ids,final_predictions]).T
submission.rename(columns={"Unnamed 0": "Overall_Experience"},inplace=True)
submission.to_csv('submission.csv',index = False)
```
